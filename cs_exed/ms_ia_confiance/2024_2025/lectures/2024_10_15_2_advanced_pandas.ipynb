```python
import numpy as np
import pandas as pd
from skrub.datasets import fetch_employee_salaries
```

```python
dataset = fetch_employee_salaries()
employees_df, salaries = dataset.X, dataset.y
```

```python
employees_df.head()
```

```python
salaries.head()
```

```python
df = pd.concat([employees_df, salaries], axis=1)
```
